```python
## Create new column for ML by converting experience to numeric
def convert_experience_to_numeric(value):
    # Handle missing values
    if pd.isna(value):
        return np.nan

    # Clean text (remove spaces, lowercase)
    value = value.lower().strip()

    # Handle special cases
    if "less" in value:
        return 0.5
    elif "more" in value:
        return 45

    # Extract numeric parts
    numbers = [int(num) for num in value.split() if num.isdigit()]

    # Handle ranges or single values
    if len(numbers) == 2:
        return (numbers[0] + numbers[1]) / 2
    elif len(numbers) == 1:
        return numbers[0]
    else:
        return np.nan
```

```python
## Know we covert and apply convert it into numeric
df['experience_years_numeric'] = df['experience_years'].apply(convert_experience_to_numeric)
```

```python
import re
import numpy as np
import pandas as pd

def clean_salary(value):
    if pd.isna(value):
        return np.nan

    # Convert to string
    value = str(value).lower()

    # Remove currency symbols and commas
    value = re.sub(r'[\$,]', '', value)
    value = value.replace(',', '').replace('usd', '').replace('cad', '').replace('aud', '')

    # Replace common letter typos (O  0)
    value = value.replace('o', '0')

    # Extract first number found in string
    match = re.search(r'\d+', value)
    if match:
        return float(match.group())
    else:
        return np.nan

# Apply to column
df['annual_salary_ml'] = df['annual_salary'].apply(clean_salary)

# Check few rows
print(df[['annual_salary', 'annual_salary_ml']].head(15))
```

```python
# === FULL FEATURE ENGINEERING PIPELINE (run on your df) ===

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler

# ---------- 0. Quick safety checks ----------
assert 'job_title' in df.columns, "DataFrame must contain 'job_title' column."
print("Initial shape:", df.shape)

# Keep a working copy
df_fe = df.copy()

# ---------- 1. Ensure cleaned title exists ----------
if 'job_title_clean' not in df_fe.columns:
    df_fe['job_title_clean'] = df_fe['job_title'].astype(str).str.lower().str.strip()
    df_fe['job_title_clean'] = df_fe['job_title_clean'].str.replace(r'[_\-/\(\)\.]', ' ', regex=True)
    df_fe['job_title_clean'] = df_fe['job_title_clean'].str.replace(r'\s+', ' ', regex=True).str.strip()

# ---------- 2. Improved job_family mapping (reduce "Other") ----------
def improved_job_family(t):
    if pd.isna(t): return 'Other'
    s = t.lower()
    # Tech / IT
    tech_kw = ['software','developer','devops','engineer','programmer','backend','frontend','sde','qa','site reliability','sysadmin','network','it ', 'architect','dba','cloud','linux','full stack','fullstack','stack']
    if any(k in s for k in tech_kw): return 'Tech / IT'
    # Data & Analytics
    data_kw = ['data ','data scientist','data analyst','machine learning','ml ','ai ','analyst','bi ','business intelligence','statistician','dataengineer','data engineer']
    if any(k in s for k in data_kw): return 'Data & Analytics'
    # Finance
    finance_kw = ['accountant','finance','financial','cfo','auditor','controller','tax','treasury','investment','banker']
    if any(k in s for k in finance_kw): return 'Finance & Accounting'
    # Healthcare
    health_kw = ['nurse','physician','doctor','medical','pharmacist','clinic','md','rn','therapist','physician','surgeon','hospital']
    if any(k in s for k in health_kw): return 'Healthcare'
    # Education
    edu_kw = ['teacher','professor','lecturer','instructor','teaching','school','academic']
    if any(k in s for k in edu_kw): return 'Education'
    # HR / Admin
    hr_kw = ['hr ','human resources','recruiter','recruiting','administrative assistant','administrative','office manager','admin ']
    if any(k in s for k in hr_kw): return 'HR & Admin'
    # Marketing / Sales
    mkt_kw = ['marketing','social media','seo','content','brand','sales','business development','account executive','account manager','bd ']
    if any(k in s for k in mkt_kw): return 'Marketing & Sales'
    # Legal
    legal_kw = ['lawyer','attorney','paralegal','counsel','legal']
    if any(k in s for k in legal_kw): return 'Legal'
    # Management keywords fallback (map many manager/director-like to Management if not matched above)
    mgmt_kw = ['manager','director','vp ','vice president','chief','head of','coo','ceo','cxo','president','lead ']
    if any(k in s for k in mgmt_kw): return 'Management & Leadership'
    # Research / Science -> try to map to Data or Other
    if any(k in s for k in ['research','scientist','postdoc','lab','clinical']):
        if 'data' in s or 'scientist' in s: return 'Data & Analytics'
        return 'Other'
    # common role words that often belong to Tech/Data/Finance/Sales
    if any(k in s for k in ['analyst','consultant','engineer','developer','specialist','technician']):
        # fallback heuristics
        if 'data' in s or 'analytic' in s or 'business intelligence' in s:
            return 'Data & Analytics'
        if 'engineer' in s or 'developer' in s or 'tech' in s:
            return 'Tech / IT'
        if 'account' in s and 'manager' not in s:
            return 'Finance & Accounting'
    # fallback
    return 'Other'

df_fe['job_family2'] = df_fe['job_title_clean'].apply(improved_job_family)

# ---------- 3. Seniority cleanup: merge rare into sensible buckets ----------
def extract_seniority_simple(t):
    if pd.isna(t) or t.strip()=='' or t.lower() in ['none','nan','other']: return 'mid'
    s = t.lower()
    if any(k in s for k in ['intern','internship']): return 'junior'
    if any(k in s for k in ['junior','jr ','jr.','assistant']): return 'junior'
    if any(k in s for k in ['senior','sr ','sr.','sr-']): return 'senior'
    if any(k in s for k in ['lead ','lead.','principal','head','staff']): return 'senior'
    if any(k in s for k in ['manager','mgr','supervisor']): return 'manager'
    if any(k in s for k in ['director','vp ','vice president','chief ','cfo','cto','ceo','president']): return 'executive'
    return 'mid'

# If job_seniority column already exists use it, else build from job_title_clean
if 'job_seniority' not in df_fe.columns:
    df_fe['job_seniority'] = df_fe['job_title_clean'].apply(extract_seniority_simple)
else:
    # map existing to cleaned
    df_fe['job_seniority'] = df_fe['job_seniority'].astype(str).apply(lambda x: extract_seniority_simple(x))

# Merge very rare seniority levels (done in extract function above)
df_fe['job_seniority'] = df_fe['job_seniority'].replace({'lead':'senior'})

# ---------- 4. Country grouping (simple) ----------
# lower-case column
if 'country_ml' in df_fe.columns:
    c = df_fe['country_ml'].astype(str).str.lower()
    def country_group(x):
        if pd.isna(x) or x == 'nan' or x.strip()=='':
            return 'Unknown'
        if 'united states' in x or x=='usa' or x=='us' or x.strip()=='none' or x.strip()== 'none':
            return 'United States'
        if 'canada' in x or x=='ca' or 'ontario' in x:
            return 'Canada'
        if 'uk' in x or 'united kingdom' in x or 'england' in x:
            return 'United Kingdom'
        if 'australia' in x or x=='aus' or 'nz' in x or 'new zealand' in x:
            # combine Australia & NZ into ANZ (optional)
            return 'Australia'
        if any(k in x for k in ['germany','france','spain','italy','nether','poland','sweden','norway','denmark']):
            return 'Europe'
        if any(k in x for k in ['india','china','pakistan','philippines','japan','singapore','south korea']):
            return 'Asia'
        if 'mexico' in x or 'brazil' in x or 'argentina' in x:
            return 'LATAM'
        return 'Other'
    df_fe['country_group'] = df_fe['country_ml'].astype(str).str.lower().apply(country_group)
else:
    df_fe['country_group'] = 'Unknown'

# ---------- 5. Industry coarse grouping ----------
def industry_group(x):
    if pd.isna(x): return 'Other'
    s = str(x).lower()
    if any(k in s for k in ['tech','software','it','information technology','computer','internet']):
        return 'Tech'
    if any(k in s for k in ['finance','bank','account','investment','insurance']):
        return 'Finance'
    if any(k in s for k in ['health','medical','hospital','clinic','pharma']):
        return 'Healthcare'
    if any(k in s for k in ['education','school','university','college']):
        return 'Education'
    if any(k in s for k in ['nonprofit','non profit','non-profit']):
        return 'Nonprofit'
    if any(k in s for k in ['retail','ecommerce','consumer']):
        return 'Retail'
    if any(k in s for k in ['manufactur','industrial','factory']):
        return 'Manufacturing'
    if any(k in s for k in ['government']):
        return 'Government'
    if any(k in s for k in ['marketing','media','advertis']):
        return 'Marketing'
    return 'Other'

df_fe['industry_group'] = df_fe['industry'].apply(industry_group)

# ---------- 6. Derived numeric features ----------
# experience per age (add small eps)
df_fe['exp_per_age'] = df_fe['experience_years_numeric'] / (df_fe['age_num'] + 1e-6)

# simple bins (optional)
df_fe['age_bucket'] = pd.cut(df_fe['age_num'], bins=[15,25,35,45,55,100], labels=['<25','25-35','35-45','45-55','55+'])
df_fe['exp_bucket'] = pd.cut(df_fe['experience_years_numeric'], bins=[0,2,5,10,20,50], labels=['0-2','2-5','5-10','10-20','20+'])

# ---------- 7. Train / Test split (compute mean encodings on train only!) ----------
TARGET = 'salary_final'   # using salary_final; change if you want log salary
X = df_fe.drop(columns=[TARGET])
y = df_fe[TARGET]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# ---------- 8. Compute train-based mean encodings (no leakage) ----------
# industry mean salary
industry_mean = y_train.groupby(X_train['industry_group']).mean().to_dict()
X_train['industry_mean_salary'] = X_train['industry_group'].map(industry_mean)
X_test['industry_mean_salary']  = X_test['industry_group'].map(industry_mean).fillna(y_train.median())

# job_family mean salary (use refined job_family2)
jobfam_mean = y_train.groupby(X_train['job_family2']).mean().to_dict()
X_train['jobfam_mean_salary'] = X_train['job_family2'].map(jobfam_mean)
X_test['jobfam_mean_salary']  = X_test['job_family2'].map(jobfam_mean).fillna(y_train.median())

# country_group mean salary
country_mean = y_train.groupby(X_train['country_group']).mean().to_dict()
X_train['country_mean_salary'] = X_train['country_group'].map(country_mean)
X_test['country_mean_salary']  = X_test['country_group'].map(country_mean).fillna(y_train.median())

# frequency encoding for job_title_clean (train counts)
job_counts = X_train['job_title_clean'].value_counts().to_dict()
X_train['job_title_freq'] = X_train['job_title_clean'].map(job_counts)
X_test['job_title_freq'] = X_test['job_title_clean'].map(job_counts).fillna(0)

# optionally make a "is_top_job_title" binary for very common roles
top_titles = set(X_train['job_title_clean'].value_counts().index[:200])  # top 200 keepers
X_train['job_title_top'] = X_train['job_title_clean'].apply(lambda x: 1 if x in top_titles else 0)
X_test['job_title_top']  = X_test['job_title_clean'].apply(lambda x: 1 if x in top_titles else 0)

# ---------- 9. Encoding: ordinal for seniority, one-hot for small groups ----------
# seniority ordinal map
sen_map = {'junior':0, 'mid':1, 'senior':2, 'manager':3, 'executive':4}
X_train['seniority_code'] = X_train['job_seniority'].map(sen_map).fillna(1)
X_test['seniority_code'] = X_test['job_seniority'].map(sen_map).fillna(1)

# one-hot small card columns: job_family2 (should now be small), industry_group, country_group, age/exp buckets
ohe_cols = ['job_family2','industry_group','country_group','age_bucket','exp_bucket']
X_train_ohe = pd.get_dummies(X_train[ohe_cols], drop_first=True)
X_test_ohe  = pd.get_dummies(X_test[ohe_cols], drop_first=True)

# Align columns
X_train_ohe, X_test_ohe = X_train_ohe.align(X_test_ohe, join='left', axis=1, fill_value=0)

# ---------- 10. Build final numeric matrix for models ----------
numeric_cols = ['age_num','experience_years_numeric','exp_per_age','job_title_freq',
                'jobfam_mean_salary','industry_mean_salary','country_mean_salary','job_title_top','seniority_code']

# Fill NaNs in numeric_cols
X_train[numeric_cols] = X_train[numeric_cols].fillna(0)
X_test[numeric_cols]  = X_test[numeric_cols].fillna(0)

X_train_final = pd.concat([X_train[numeric_cols].reset_index(drop=True), X_train_ohe.reset_index(drop=True)], axis=1)
X_test_final  = pd.concat([X_test[numeric_cols].reset_index(drop=True),  X_test_ohe.reset_index(drop=True)], axis=1)

# Align final
X_train_final, X_test_final = X_train_final.align(X_test_final, join='left', axis=1, fill_value=0)

print("Final shapes:", X_train_final.shape, X_test_final.shape)

# ---------- 11. Quick scaling for Ridge (only numeric part) ----------
scaler = StandardScaler()
Xtr_scaled = scaler.fit_transform(X_train_final)   # scale everything; tree models don't need scaling but ridge does
Xte_scaled = scaler.transform(X_test_final)

# ---------- 12. Quick baseline models ----------
def evaluate_model(y_true, y_pred, name="Model"):
    print(f"--- {name} ---")
    print("R2:", r2_score(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print()

# Random Forest (unscaled features)
rf = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42, n_jobs=-1)
rf.fit(X_train_final, y_train)
pred_rf = rf.predict(X_test_final)
evaluate_model(y_test, pred_rf, "RandomForest")

# Ridge (scaled)
ridge = Ridge(alpha=1.0)
ridge.fit(Xtr_scaled, y_train)
pred_ridge = ridge.predict(Xte_scaled)
evaluate_model(y_test, pred_ridge, "Ridge")

# ---------- 13. Quick feature importances (RF) ----------
importances = rf.feature_importances_
feat_names = X_train_final.columns
feat_imp = pd.Series(importances, index=feat_names).sort_values(ascending=False).head(30)
print("Top features from RandomForest:")
print(feat_imp)
```
